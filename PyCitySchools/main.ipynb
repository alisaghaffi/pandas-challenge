```python

# Dependencies and Setup 
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "C:/Users/alisa/GIT/pandas-challenge/PyCitySchools/schools_complete.csv"
student_data_to_load = "C:/Users/alisa/GIT/pandas-challenge/PyCitySchools/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#----------------------------School Summary----------------------

#school name
#school_name = school_data['school_name']
#print (school_name)

#School Type
#school_type = school_data['type']
#print (school_type)


#school_type
school_type = school_data.set_index(["school_name"])["type"]

#Total Students size
#Total_Students = school_data['size']
Total_Students = school_data.set_index(["school_name"])["size"]

total_number_of_students  = school_data_complete["school_name"].value_counts()

#print (Total_Students)

#Total School Budget
Total_School_Budget = round(school_data.set_index(["school_name"])["budget"],3)

#print (Total_School_Budget)

#Per Student Budget
Per_student_Budget = Total_School_Budget / Total_Students
#print(Per_student_Budget)


total_number_of_students  = school_data_complete.groupby(["school_name"]).count()["student_name"]

#Average Math Score
average_math_score  = school_data_complete.groupby(["school_name"]).mean()["math_score"]

#print(average_math_score)

#Average Reading Score
average_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
#print(average_reading_score)

#% Passing Math
students_with_passing_math_score = school_data_complete[(school_data_complete["math_score"] >= 70)].count()
#print(students_with_passing_math_score[1])
percent_math_passing =  round(students_with_passing_math_score[1] / Total_Students,8)*100
#print(percent_math_passing)


#Passing Reading
students_with_passing_Reading_score = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()
percent_Reading_passing =  round(students_with_passing_Reading_score[1] / Total_Students,8)*100
#print(percent_Reading_passing)

#Overall Passing (The percentage of students that passed math **and** reading.)
Overall_Passing = school_data_complete[school_data_complete['math_score'and'reading_score'] >=70].count()

percent_Overall_Passing =  round(Overall_Passing[1] / Total_Students,8)*100 

#percent_Overall_Passing

#df = pd.DataFrame(school_type)
#print(df)

school_summary = pd.DataFrame(
    {
        "School Type": school_type,
         "Total Students": Total_Students,
         "Total School Budget": Total_School_Budget,
         "Per Student Budget": Per_student_Budget,
       "Average Math Score": average_math_score,
        "Average Reading Score": average_reading_score,
       "% Passing Math": percent_math_passing,
        "% Passing Reading": percent_Reading_passing,
      "% Overall Passing": percent_Overall_Passing,


        
    }
)
school_summary.head()
school_summary_sorted = school_summary.sort_values(by=['school_name'], inplace=False, ascending=True)
#school_summary_sorted = school_summary.sort_values(by=['school_name'], inplace=False, ascending=True)
school_summary_sorted.head(15)

```

```python
#--------------------------Scores by School Spending------------------------
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending.
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$584", "$585-629", "$630-644", "$645-680"]

Per_student_Budget.describe()

```

```python
pd.cut(Per_student_Budget, spending_bins)
```

```python
Per_student_Budget.groupby(pd.cut(Per_student_Budget, spending_bins)).count()
```

```python
school_summary["Spending lable"] = pd.cut(Per_student_Budget, spending_bins, labels=group_names)
school_summary
```

```python
spending_math_scores = school_summary.groupby(["Spending lable"]).mean()["Average Math Score"]

spending_reading_scores = school_summary.groupby(["Spending lable"]).mean()["Average Reading Score"]

spending_passing_math = school_summary.groupby(["Spending lable"]).mean()["% Passing Math"]

spending_passing_reading = school_summary.groupby(["Spending lable"]).mean()["% Passing Reading"]

overall_passing_spending = school_summary.groupby(["Spending lable"]).mean()["% Overall Passing"]

#print(overall_passing_spending)
```

```python
spending_Budget_summary_student = pd.DataFrame(
    {
        "Average Math Score": spending_math_scores,
        "Average Reading Score": spending_reading_scores,
        "% Passing Math": spending_passing_math,
        "% Passing Reading": spending_passing_reading,
        "% Overall Passing": overall_passing_spending,
    }
)
spending_Budget_summary_student

spending_Budget_summary_student.head().to_html('C:/Users/alisa/GIT/pandas-challenge/PyCitySchools/spending_Budget_summary_student.html')
spending_Budget_summary_student.head().to_csv('C:/Users/alisa/GIT/pandas-challenge/PyCitySchools/spending_Budget_summary_student.csv')

spending_Budget_summary_student
```
